```python
# Connect to Mongo database
client = MongoClient('localhost', 27017)
db = client['ouroboros_staging']
subjects = db['radio_subjects']
classifications = db['radio_classifications']
```

```python
# Query database and create panda dataframe
batch = classifications.find({"updated_at": {"$gt": beta_release_date}})
df = pd.DataFrame(list(batch))
```

```python
# Get classification count per user
user_name = df.user_name
volunteers = pd.value_counts(df.user_name)

# Get classification counts
total_count = df._id.count()
anonymous_count = total_count - df.user_name.count()

volunteers = volunteers.set_value("anonymous", anonymous_count);
volunteers.sort(ascending=False)
```

```python
# Plot classification count per user
volunteers.plot(kind='bar', figsize=(18, 6))
```

```python
# Get user agents
f = lambda x: x[-1]["user_agent"]
user_agents = pd.value_counts( df.annotations.map(f) )
user_agents.plot(kind='bar', figsize=(18, 6))
```

```python
# Ensure subject classification is uniform
f = lambda x: x[0]["id"]
subjects = pd.value_counts( df.subjects.map(f) )
subjects.plot(logy=True, figsize=(18, 6))
```

```python
# Get number of radio identifications over classifications
def f(x):
    if x[0].has_key("0"):
        a = x[0]["0"]
        if a.has_key("radio"):
            return len(a["radio"])
pd.value_counts( df.annotations.map(f) ).plot(kind='bar', figsize=(18, 6) )
```

```python
# Get number of IR identifications over classifications
def f(x):
    if x[0].has_key("0"):
        a = x[0]["0"]
        if a.has_key("infrared"):
            return len(a["infrared"])
pd.value_counts( df.annotations.map(f) ).plot(kind='bar', figsize=(18, 6) )
```
